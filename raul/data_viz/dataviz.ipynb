```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle("./train_cleaned.pkl")
df.columns
```

```python
fig1, ax1 = plt.subplots(1, 1, figsize=(15, 15))
sns.histplot(data=df['travel_time'], bins=250, ax=ax1)
ax1.set_xlabel("Total Drive Time (Seconds)", size=18)
ax1.set_ylabel("Count", size=18)
```

```python
fig2, ax2 = plt.subplots(1, 1, figsize=(15, 15))
sns.scatterplot(data=df[['TAXI_ID', 'travel_time']], x='TAXI_ID', y='travel_time', ax=ax2)
ax2.set_title('Travel Time for Taxi Drivers')
```
