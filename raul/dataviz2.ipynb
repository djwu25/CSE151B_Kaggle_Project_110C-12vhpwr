```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./cse151b_data/metaData_taxistandsID_name_GPSlocation.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df = df.dropna(axis=0)
```

```python
df.isnull().any().sum()
```

```python
df['Latitude'] = df['Latitude'].apply(lambda x: float(x))
```

```python
fig1, ax1 = plt.subplots(1, 1, figsize=(15, 15))
sns.kdeplot(data=df[['Latitude', 'Longitude']], x='Latitude', y='Longitude', ax=ax1, fill=True, cmap='mako', thresh=0)
ax1.set_xlabel('Latitude', size=18)
ax1.set_ylabel('Longitude', size=18)
```
